```python
import pandas as pd
from bokeh.plotting import figure, show, output_file
```

```python
f=pd.read_excel("verlegenhuken.xlsx",sheet_name=0)
```

```python
temperature=[i/10 for i in f["Temperature"]]
pressure=[p/10 for p in f["Pressure"]]
```

```python
p.title.text="Temperature and Air Pressure"
p.title.text_font="arial"
p.title.text_color="gray"

p.title.text_font_style="bold"

p.xaxis.axis_label="Temperature (Degree Celsius)"
p.yaxis.axis_label="Pressure(hPa)"

p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
```

```python
p.circle(temperature,pressure, color="blue", size=0.5, alpha=2.0)
```

```python
output_file("Weather.html")
show(p)
```
